```
df=spark.read.csv('/FileStore/tables/wc.csv', sep=',',header=True,inferSchema=True)
```

```
df.show()
```

```
data=df.withColumn('Height',df['Height']+1)
data=data.withColumn('Name',data['FIFA Popular Name'])
data=data.select('Name','Height')
data.show(20)
```

```
df.filter(df['Height']>170).show()
```

```
from pyspark.sql.functions import when
data = df.withColumn("result", \
              when(df["Height"]> 170, 1).otherwise(0))
data.show(100)
```

```
from pyspark.sql.functions import max, mean, min, corr, stddev, kurtosis, skewness
```

```
a=df.select(min('Height'))
a=(a.select('min(Height)').
      rdd.flatMap(lambda x: x).collect())[0]
df.filter(df['Height']==a).select('FIFA Popular Name','Height').show()
```

```
a=df.select(max('Height'))
a=(a.select('max(Height)').
      rdd.flatMap(lambda x: x).collect())[0]
df.filter(df['Height']==a).show()
```

```
df.filter(df['Team']=='Argentina').select(mean('Height')).show()
```
